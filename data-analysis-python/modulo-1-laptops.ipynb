```python
# importamos librerias que usaremos
import pandas as pd
import numpy as np
```

```python
# url donde se encuentran los datos
file_path ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
```

```python
# cargamos el dataset en un DataFrame de pandas
df = pd.read_csv(file_path)
```

```python
# vemos las primeras 5 filas para confirmar la carga
df.head()
```

```python
# creamos una lista con los encabezados
headers = [ "Manufacturer", "Category", "Screen", "GPU", "OS", "CPU_core",
            "Screen_Size_inch", "CPU_frequency", "RAM_GB", "Storage_GB_SSD",
            "Weight_kg","Price"]
```

```python
# los reemplazamos e imprimimos para verificar
df.columns = headers
df.head(10)
```

```python
df.tail(10)
```

```python
# reemplazamos los valores '?' con NaN para poder eliminarlos
df1 = df.replace('?', np.NaN)
```

```python
# filas completas que tienen al menos un valor NaN
print(df1[df1.isna().any(axis=1)])

# Si buscamos en una columna específica:
# print(df[df['nombre_columna'].isna()])
```

```python
# eliminamos los valores faltantes de Weight_kg y Screen_Size_inch
df = df1.dropna(subset=['Weight_kg','Screen_Size_inch'], axis=0)
```

```python
# verificamos
df[df[['Weight_kg', 'Screen_Size_inch']].isna().any(axis=1)]
```

```python
# filas completas que tienen al menos un valor NaN
print(df[df.isna().any(axis=1)])
```
