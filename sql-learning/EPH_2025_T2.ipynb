#### Aclaraciones metodológicas:

| Ponderador | Aplica a Variables | ¿Qué Mide o Cuenta? |
|---|---|---|
| PONDER | Todo lo que NO es ingreso (edad, ocupación, sexo, vivienda) | La población total o cantidad de hogares por características no monetarias |
| PONDII | Ingreso Total Individual (`P47T`) y no laborales | El ingreso total de las personas |
| PONDIIO | Ingreso de la Ocupación Principal (`P21`) | El ingreso de las personas solo por su trabajo principal |
| PONDIH | Ingreso Total Familiar (`ITF`, `IPCF`) | La población sujeta a análisis de pobreza y distribución del ingreso (del hogar) |

#### Cantidad de muestras realizadas por hogares en el Aglomerado N° 93 (Viedma-Carmen de Patagones)

```python
%%sql

-- Contar los hogares entrevistados

SELECT COUNT(*)
FROM hogar
WHERE AGLOMERADO = 93;
```

#### ¿Cuál es el número total estimado de hogares en el aglomerado urbano Viedma-Patagones?

```python
%sql SELECT SUM(PONDERA) \
    FROM hogar \
    WHERE AGLOMERADO =93;
```

#### De numero total estimado de hogares en el alomerado urbano Viedma-Patagones, ¿cuántos de esos pertenece a familias propitarias y cuántas a familias inquilinas?

```python
%%sql

SELECT 
    SUM(CASE WHEN II7 IN (1, 2) THEN PONDIH ELSE 0 END) AS hogares_propietarios_estimados,
    SUM(CASE WHEN II7 = 3 THEN PONDIH ELSE 0 END) AS hogares_inquilinos_estimados
FROM hogar
WHERE AGLOMERADO = 93;
```

### ¿Cuántas familias estimadas perciben un ingreso mayor o igual a 3 000 000 de pesos?

```python
%%sql

SELECT SUM(PONDIH)
FROM hogar
WHERE AGLOMERADO = 93
AND ITF >= 3000000;
```

#### De lo que perciben un ingreso superior a los 3 000 000 millones de pesos, ¿cuántas familias son propietadas?

```python
%%sql

SELECT SUM(PONDIH) AS familias_propietarias_altos_ingresos
FROM hogar
WHERE AGLOMERADO = 93 AND ITF >=3000000 AND II7 IN (1,2);
```

## Resultados parciales

* Coincide que, aquellas familias que tienen un ingreso considerado "superior" al momento del análisis, son las mismas familias que son propietarias. 

# Familias pobres

### Referencia CBT (Canasta Básica Total)

Para un análisis con rigor regional en el Aglomerado 93 (Viedma-Carmen de Patagones, Región Patagonia), se utiliza la CBT ajustada:

-  **Fuente Oficial:** **INDEC** - Valorización mensual de la Canasta Básica Total (Informe de Junio de 2025).
-  **Link de referencia:** [https://www.indec.gob.ar/uploads/informesdeprensa/canasta\_07\_25B95CBBB21C.pdf](https://www.indec.gob.ar/uploads/informesdeprensa/canasta_07_25B95CBBB21C.pdf)
- **Metodología de Ajuste (4 Personas):** El valor es el resultado de multiplicar la CBT del Adulto Equivalente de la Patagonia por **3.09** (el total de Unidades Consumidoras para este tipo de hogar).
- **Valor de Corte Utilizado (CBT H-4 Patagonia):** **\$1.128.895**

#### De los 34323 hogares estimados en el anglomerado urbano de Vieda-Patagones, ¿cuál es el porcentaje que se encuentran por debajo de la línea de pobreza (teniendo en cuenta la CBT)?


```python
%%sql
SELECT
    -- Numerador: Suma de HOGARES de 4 personas que son pobres (ITF < CBT_H4)
    (SUM(CASE WHEN T1.ITF < 1128895 THEN T1.PONDIH ELSE 0 END) * 100.0)
    /
    -- Denominador: Suma del total de HOGARES de 4 personas en la muestra
    SUM(T1.PONDIH) AS Tasa_Pobreza_Hogares_4_Personas_Estimada
FROM
    hogar AS T1
WHERE
    T1.AGLOMERADO = 93
    AND T1.IX_TOT = 4
    AND T1.ITF >= 0;
```

## Conclusiones preliminares

* Se estima que: aproximadamente el 41.44 % de los hogares de 4 personas en el aglomerado Viedma-Carmen de Patagones se encuentran por debajo de la línea de pobreza.
